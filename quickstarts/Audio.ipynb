### Exemplo DEF 01 - Passagem de parâmetro por Valor.

```
#Declaração da funcão votante()
def votante():
  idade = int(input("Informe a idade do eleitor: "))

  if idade >=18 and idade<=70:
    print(f"Idade do eleitor: {idade} - Voto obrigatório!")
  elif idade<16:
    print(f"Idade do eleitor: {idade} - Voto inválido!")
  else:
    print(f"Idade do eleitor: {idade} - Voto facultativo!")

#Execução à partir daqui:
resposta = True
while (resposta == True):
  votante()

  prox_eleitor = input("Próximo votante? - Digite (SIM) ou (NÃO)")

  if prox_eleitor != "SIM":
   resposta = False
```

### Exemplo DEF 02 - Passagem de parâmetro por Referência.

```
#Declaração da funcão votante()
def votante(idade):

  if idade >=18 and idade<=70:
    print(f"Idade do eleitor: {idade} - Voto obrigatório!")
  elif idade<16:
    print(f"Idade do eleitor: {idade} - Voto inválido!")
  else:
    print(f"Idade do eleitor: {idade} - Voto facultativo!")

#Execução à partir daqui:
votante(60) # Parâmetro passado de forma manual referente a idade.
votante(15) # Parâmetro passado de forma manual referente a idade.
votante(17) # Parâmetro passado de forma manual referente a idade.

```

### Exemplo DEF 03 - Passagem de 2 parâmetros por de referência.

```
#Declaração da funcão votante()
def votante(idade, nome):
  # idade = int(input("Informe a idade do eleitor: "))

  if idade >=18 and idade<=70:
    print(f"Idade - {idade} anos - eleitor: {nome} - Voto obrigatório!")
  elif idade<16:
    print(f"Idade - {idade} anos - eleitor: {nome} - Voto inválido!")
  else:
    print(f"Idade - {idade} anos - eleitor: {nome} - Voto facultativo!")

#Execução à partir daqui:
votante(60, "Pedro") # Parâmetro passado de forma manual referente a idade e nome.
votante(15, "Kelle") # Parâmetro passado de forma manual referente a idade e nome.
votante(17, "Patricia") # Parâmetro passado de forma manual referente a idade e nome.
```
